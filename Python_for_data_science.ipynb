```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

# Load dataset
try:
    data = pd.read_csv('Retail_Sales.csv')
    print("Dataset:")
    display(data)

    # Prepare data
    X = data[['Advertising_Budget']]  # Feature
    y = data['Sales']  # Target

    # Train model
    model = LinearRegression()
    model.fit(X, y)

    # Predict for new budgets
    budgets = [[20000], [40000], [50000]]  # Test budgets
    for budget in budgets:
        # Reshape the budget to be a 2D array with one sample and one feature
        predicted_sales = model.predict([budget])[0]
        print(f"Predicted Sales for {budget[0]} PKR budget: {predicted_sales:.2f} PKR")

    # Plot
    plt.scatter(X, y, color='blue', label='Actual Sales')
    plt.plot(X, model.predict(X), color='red', label='Prediction Line')
    plt.xlabel('Advertising Budget (PKR)')
    plt.ylabel('Sales (PKR)')
    plt.title('Linear Regression: Sales vs Advertising Budget')
    plt.legend()
    plt.show()

except FileNotFoundError:
    print(" Error: CSV file not found. Please check the filename and location.")

except KeyError as e:
    print(f" Error: Column '{e}' not found in the dataset.")

except ValueError as e:
    print(f" Error: {e}. Please check the shape of your data.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
import numpy as np

# Create Student_marks.csv
try:
    data = pd.read_csv('Student_marks.csv')
except FileNotFoundError:
    with open("Student_marks.csv", "w") as f:
        f.write("Name,Marks,Pass\n")
        f.write("Ali,85,1\n")
        f.write("Sara,90,1\n")
        f.write("Ahmed,55,0\n")
        f.write("Fatima,92,1\n")
        f.write("Zain,78,1\n")
        f.write("Bilal,65,0\n")
    print("Student_marks.csv created successfully!")
    data = pd.read_csv('Student_marks.csv')

print("Dataset:")
print(data)

# Prepare data
X = data[['Marks']]  # Input (marks as 2D array)
y = data['Pass']     # Output (0 or 1)

# Train model
model = LogisticRegression()
model.fit(X, y)

# Predictions
marks = [[60], [70], [75], [82]]
for mark in marks:
    predicted_pass = model.predict(np.array(mark).reshape(1, -1))[0]
    print(f"Prediction for {mark[0]} marks: {'Pass' if predicted_pass == 1 else 'Fail'}")

# Plot
plt.figure(figsize=(8, 6))
plt.scatter(X, y, color='blue', label='Actual Pass/Fail', s=100)
plt.xlabel('Marks', fontsize=12)
plt.ylabel('Pass (1) / Fail (0)', fontsize=12)
plt.title('Logistic Regression: Pass vs Marks', fontsize=14)
plt.legend()
plt.grid(True)
plt.savefig('pass_fail_plot.png', dpi=300)  # Save for LinkedIn
plt.show()

# Accuracy score
from sklearn.metrics import accuracy_score
y_pred = model.predict(X)
print("Accuracy Score:", accuracy_score(y, y_pred))
```
